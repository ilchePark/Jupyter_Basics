# 1. Simple Linear Regression

Se refiere a una variable independiente para hacer una predicción, en el caso de Multiple Linear Regression se refiere a multiples variables independientes para hacer una predicción

Y = b0 + b1*x



```python

import pandas as pd
csv_path = 'C:\\Users\ilche_st130to\Downloads\Jupyter\datos.csv'
df= pd.read_csv(csv_path)

df.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LM = LinearRegression()
```

```python
X = df[["Edad"]]
Y = df[["Sueldo"]]
```

```python
LM.fit(X,Y)
```

```python
Yhat = LM.predict(X)
Yhat
```

```python
LM.intercept_
```

```python
LM.coef_
```

La relación entre Edad y Sueldo es:

Sueldo = (Lm.intercept_ + (Lm.coef_ * edad) ) .... Ejemplo:


2998.691187 = -3644.00728332 + (  265.70793882 * 25 ) 

Interpretación:  A los 25 años podemos predecir con respecto a la base de datos que tenemos que estarás ganando 2998.69













# 2. Multiple Linear Regression

Para predecir un número no solo basta evaluar sólo una variable, si no varias como por ejemplo:

Y = Es el objetivo.
